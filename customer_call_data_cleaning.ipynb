```python
import pandas as pd
df= pd.read_excel("/content/Customer Call List.xlsx")

df
```

```python
df = df.drop_duplicates()
df = df.drop(columns="Not_Useful_Column")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r"[\/|]", "-", regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace("N-a", "")
df["Phone_Number"] = df["Phone_Number"].fillna("")

df
```

```python
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)
df['Street_Address'] = df['Street_Address'].str.strip()
df['State'] = df['State'].str.strip()
df['Zip_Code'] = df['Zip_Code'].str.strip()

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({"Yes":"Y","No":"N"})

df["Paying Customer"] = df["Paying Customer"].replace({"Yes":"Y","No":"N"})


df
```

```python
df = df[df["Do_Not_Contact"] != "Y"]
df = df[df["Phone_Number"] != ""]

df
```

```python
df = df.reset_index(drop=True)

df.head()
```
